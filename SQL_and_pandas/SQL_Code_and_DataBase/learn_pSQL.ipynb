### Data Definition Language (DDL)
The Blueprint of SQL

```python
-- safe protocol: drop the table if it exists, to avoid errors
DROP TABLE IF EXISTS jesse_spark;
```

```python
-- creating a new table and assigning columns, each with it's own data type
CREATE TABLE jesse_spark(
    user_id SERIAL PRIMARY KEY, -- serial means that an incrementing number is placed in this column uponn the creation of a new row

    user_name VARCHAR(55), -- this means that the columns only allows text of up to 55 characters

    user_k_score FLOAT, -- decimal numbers

    is_cool BOOLEAN, -- booleans (True/False)

    join_time TIMESTAMP DEFAULT NOW() -- time is to be added to this column. 'DEFAULT NOW() means that if nothing was written for this column upon the creation of a new row, Postgres should figure out the current time and add it there'
);
```

```python
-- delete all existing data in the table 'jesse_spark', without deleting the table itself
TRUNCATE TABLE jesse_spark;
```

```python
SELECT * FROM jesse_spark;
```

### Data Manipulation Language (DML)

```python
INSERT INTO jesse_spark (user_name, is_cool)
VALUES ('Nwachukwu Jesse', FALSE)
RETURNING user_id, join_time;
```

```python
INSERT INTO jesse_spark(user_name, user_k_score, is_cool)
VALUES
('Kelly Nigga', 2.34, FALSE),
('F Nih', 4.64, TRUE),
('Joe Biden', 1.99, FALSE)
```

```python
UPDATE jesse_spark
SET user_k_score = 2.94,
is_cool = TRUE
WHERE user_name = 'Nwachukwu Jesse' AND
user_k_score IS NULL;
```

```python
DELETE FROM jesse_spark
WHERE is_cool = FALSE;
```

```python
UPDATE jesse_spark
SET user_k_score = 1.11
WHERE user_name = 'F Nih' AND is_cool = True;
```
